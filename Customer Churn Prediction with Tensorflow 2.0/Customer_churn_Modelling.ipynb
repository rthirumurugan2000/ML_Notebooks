```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import pandas as pd
```

```python
tf.__version__
```

```python
# Import the dataset
data = pd.read_csv('Customer_Churn_Modelling.csv')
```

```python
data.head()
```

```python
x= data.drop(labels=['CustomerId','RowNumber','Surname','Exited'],axis = 1)
y= data['Exited']
```

```python
x.head()
```

```python
y.head()
```

```python
# data preprocessing
from sklearn.preprocessing import LabelEncoder
```

```python
label1 = LabelEncoder()
x['Geography'] = label1.fit_transform(x['Geography'])
label2 = LabelEncoder()
x['Gender'] = label2.fit_transform(x['Gender'])
```

```python
x = pd.get_dummies(x, drop_first=True, columns=['Geography'])
x.head(10)
```

```python
#train_test_split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0, stratify = y)
```

```python
#feauture scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train,X_test
```

#### Building Artificial neural network:

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Flatten, Dense
```

```python
model = Sequential()
model.add(Dense(x.shape[1], activation='relu', input_dim = x.shape[1]))
model.add(Dense(128, activation='relu'))
model.add(Dense(1, activation = 'sigmoid'))
```

```python
x.shape
```

```python
#compiling
model.compile(optimizer='adam', loss = 'binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train.to_numpy(), batch_size = 10, epochs = 25, verbose = 1)
```

```python
y_pred = model.predict_classes(X_test)
y_pred
```

```python
y_test
```

```python
model.evaluate(X_test, y_test.to_numpy())
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
