## DATA WRANGLING

```python
titanic_train.isnull()
```

```python
titanic_train.isnull().sum()
```

```python
sns.boxplot(x="Pclass",y="Age",data = titanic_train)
```

```python
titanic_train.drop("Cabin",axis =1,inplace=True)
```

```python
titanic_train.dropna(inplace=True)
```

```python
titanic_train.isnull().sum()
```

```python
sex = pd.get_dummies(titanic_train['Sex'],drop_first=True)
```

```python
sex.head()
```

```python
embark = pd.get_dummies(titanic_train['Embarked'],drop_first=True)
```

```python
pcl = pd.get_dummies(titanic_train['Pclass'],drop_first=True)
```

```python
titanic_train = pd.concat([titanic_train,sex,embark,pcl],axis=1)
```

```python
titanic_train.drop(['PassengerId'],axis=1,inplace=True)
```

```python
titanic_train.head()
```

```python
titanic_test.drop("Cabin",axis =1,inplace=True)
```

```python
titanic_test.dropna(inplace=True)
```

```python
titanic_test.isnull().sum()
```

```python
X = titanic_train.drop('Survived',axis = 1)
```

```python
Y = titanic_train['Survived']
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
```

```python
logmodel.fit(X,Y)
```

```python
sex = pd.get_dummies(titanic_test['Sex'],drop_first=True)
```

```python
pcl = pd.get_dummies(titanic_test['Pclass'],drop_first=True)
```

```python
embark = pd.get_dummies(titanic_test['Embarked'],drop_first=True)
```

```python
titanic_test = pd.concat([titanic_test,sex,embark,pcl],axis=1)
```

```python
titanic_test.drop(['PassengerId','Sex','Embarked','Name','Ticket','Pclass'],axis=1,inplace=True)
```
